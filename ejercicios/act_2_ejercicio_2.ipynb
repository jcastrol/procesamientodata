b. Vamos a transformar este dataset en un dataset de clasificación. El objetivo
de esta clasificación será predecir si cada una de las personas que
aparecen en el dataset trabajan en el ámbito STEM o si no lo hacen.
Recuerda que STEM es el acrónimo de Science, Technology, Engineering
and Mathematics. Para crear el problema de clasificación definiremos una
variable nueva en el DataFrame df_red llamada “is_stem”. Utiliza el método
apply de Python, combinado con una función lambda, para crear la
variable “is_stem” de forma que esta variable valga 1 si la persona trabaja
en estos ámbitos y valga 0 si no trabaja en estos ámbitos. Considerar como
valor 0 los valores faltantes o los casos en los que no se han aportado
respuesta. Eliminar la variable “job” una vez se haya creado la variable
“is_stem”.

```python
validate=lambda e: 'science' in e.lower() or 'computer'in e.lower() or 'hardware'in e.lower() or 'software'in e.lower() or 'tech'in e.lower() or 'engineering'in e.lower() or 'mathematics'in e.lower() 
df_red['is_stem']=df_red.apply(lambda e:1 if isinstance(e['job'], str) and validate(e['job']) else 0, axis = 1)
df_red.sample(10)
```

Calcula, para el conjunto de entrenamiento y para el conjunto de test, las distribuciones de las variables “age”, “orientation”, “body_type”, “height” e “income”. ¿Son las distribuciones del conjunto de test iguales a las del conjunto de entrenamiento en todos los casos? Representa las distribuciones con histogramas o gráfico de barras, según el tipo de variable. Cada histograma o gráfico de barras debe corresponder a una variable, y se deben representar las distribuciones de test y de entrenamiento. Para hacer más legible la representación, representa las distribuciones con un valor de alpha (transparencia) de 0.7. ¿Hace falta que las dos distribuciones sean iguales en todos los casos? ¿Qué consideraciones deberemos tomar para no obtener métricas distorsionadas respecto de la realidad? Tened en cuenta que, al tener los conjuntos de entrenamiento tamaños distintos, tanto los histogramas como los gráficos de barras deben estar normalizados. ¿Qué ocurre cuando se representan sin normalizar?

Calcular las distribuciones de las variables

```python
variables = ["age", "orientation", "body_type", "height", "income"]

train_distribution = {}
test_distribution = {}

for var in variables:
    train_distribution[var] = X_train[var].value_counts(normalize=True)
    test_distribution[var] = X_test[var].value_counts(normalize=True)   
```

```python
import matplotlib.pyplot as plt
```

Visualizar con histogramas 

```python
#plt.style.use('seaborn')
alpha_value = 0.5  # Transparencia para comparar mejor

# Crear gráficos para cada variable
for var in variables:
    plt.figure(figsize=(10, 6))
    # Histograma para variables numéricas
    if X_train[var].dtype in ['int64', 'float64']:
        plt.hist(X_train[var], bins=20, alpha=alpha_value, label='Entrenamiento', density=True)
        plt.hist(X_test[var], bins=20, alpha=alpha_value, label='Test', density=True)
    else:
        train_distribution[var].plot(kind='bar', alpha=alpha_value, label='Entrenamiento')
        test_distribution[var].plot(kind='bar', alpha=alpha_value, label='Test',color='orange')
    
    plt.title(f"Distribución de {var}")
    plt.xlabel(var)
    plt.ylabel("Frecuencia Normalizada")
    plt.legend()
    plt.show()
```

### **¿Hace falta que las dos distribuciones sean iguales en todos los casos?**  
No, **no es necesario que las distribuciones sean exactamente iguales en todos los casos**, pero sí es importante que mantengan una **proporción representativa** de las clases y valores del conjunto de datos original, especialmente en problemas de clasificación. Esto asegura que el modelo pueda generalizar adecuadamente y no esté sesgado hacia una clase o característica en particular.

---

### **¿Qué consideraciones deberemos tomar para no obtener métricas distorsionadas respecto de la realidad?**  
- **Estratificación:** Al dividir los conjuntos de datos, la técnica de muestreo estratificado asegura que cada subconjunto (train y test) mantenga la misma proporción de clases presentes en el conjunto original.  
- **Tamaño de los subconjuntos:** Un conjunto de entrenamiento significativamente más grande que el de test es normal, pero ambos deben reflejar la misma distribución de características.  
- **Normalización:** Dado que los conjuntos de entrenamiento y test pueden tener tamaños diferentes, es fundamental normalizar los histogramas y gráficos de barras usando frecuencias relativas (`density=True` en gráficos continuos y `value_counts(normalize=True)` en gráficos categóricos).  
- **Evitar datos desbalanceados:** Un desbalance en la representación de clases podría hacer que el modelo no aprenda correctamente a predecir clases minoritarias.

---

### **¿Qué ocurre cuando se representan sin normalizar?**  
- **Distorsión Visual:** Los conjuntos de mayor tamaño dominarán visualmente los gráficos, dificultando la comparación entre distribuciones.  
- **Comparación Incorrecta:** Se podría interpretar erróneamente que una clase es más relevante solo porque aparece más veces, en lugar de considerar su proporción real en el dataset.  
- **Impacto en Métricas:** Al entrenar un modelo con un dataset desbalanceado, las métricas de desempeño como *accuracy* podrían ser engañosas al favorecer la clase mayoritaria.
